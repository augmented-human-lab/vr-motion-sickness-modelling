# Motion Sickness Ratings

```python
import os
def CreateFolder(folder):
    if not os.path.exists(folder):
        os.mkdir(folder)
```

```python
import pandas as pd
import numpy as np
root = '../DataVersion2'
allfiles = []
for dir,sub,files in os.walk(root):
    for file in files:
        full_path = dir+'/'+file
        if '.DS_Store' in full_path: continue
        if '.ipynb' in full_path: continue
        allfiles.append(full_path)
# print(allfiles)     

def GetFrame(item,timestamp):
    # print(item)
    df = pd.read_csv(item)
    times_col = df['timestamp']
    # print(timestamp)
    # print(times_col)
    row_num = np.argmin(np.abs(times_col-timestamp))
    # print(row_num)
    frames_col = df['framecounter']
    return frames_col[row_num]
    
    
def GetFrameNum(timestamp,csvfile):
    framenum = 0
    # print(csvfile)
    selfreport_filename = csvfile.split('/')[-1].replace('.csv','')
    selfreport_filename_split = selfreport_filename.split(' ')
    gamename_sub = selfreport_filename_split[2]
    participant = selfreport_filename_split[1]
    for item in allfiles:
        if gamename_sub in item:
            if participant in item:
                if 'control.csv' in item:
                    framenum = GetFrame(item,timestamp)
                    break
    return framenum
```

```python
import csv
import pandas as pd
import numpy as np
import json
def GetMotionSicknessScore(csvfile,game):
    scores = []
    timestamps = []
    frames = []
    
    flines = open(csvfile,'r').readlines()
    for line in flines:
        if 'auto' in line.split(',')[-1]: continue
        # print(line)
        scores.append(int(line.split(',')[1]))
        timestamp_raw = int(float(line.split(',')[0]))
        timestamp = int(timestamp_raw/1000)
        # print((timestamp))
        frame = GetFrameNum(timestamp,csvfile)
        timestamps.append(timestamp)
        frames.append(frame)
    
    return scores,timestamps,frames
```

```python
import glob as glob
import os
import matplotlib.pylab as plt
mother = '../SelfReportData'
games = os.listdir('../DataVersion2')
if '.DS_Store' in games:
    games.remove('.DS_Store')
games = ['Beat Saber','Cartoon Network','Job Simulator', 'Monster Awakens', 'Traffic Cop', 'Voxel Shot','VR Rome','Epic Roller Coasters','Mini Motor Racing', 'Pottery']
shapes = ['o','*','^','x','s','+']
plt.figure(figsize=(20,10))
gameind = 1
jsondict = {}
for game in games:   
    print(game)
    
    jsonfile = 'DataExtractFolder/MotionSickness/'+game+'.json'
    game_scores = []
    plt.subplot(2,5,gameind)
    i=0
    participants = []
    for report in glob.glob(mother+'/*.csv'):
        if game not in report: continue
        # print(report)
        scores,timestamps,frames = GetMotionSicknessScore(report,game)
        plt.plot(frames,scores,shapes[i]+'-')
        game_scores = game_scores+scores
        participants.append(report.split('/')[-1].split(' ')[0])
        i+=1
    plt.title(game)
    # plt.ylabel('Motion Sickness Level')
    # plt.xlabel('Reporting Instances')
    plt.ylim(0,5)
    plt.legend(participants)
    # print(game_scores)
    print(round(np.median(game_scores),2),round(np.mean(game_scores),2),round(np.std(game_scores),2),max(game_scores))
    jsondict[game] = game_scores
    # json_string = json.dumps(jsondict, sort_keys=True, indent=4)
    # with open(jsonfile, 'w') as outfile:
    #     outfile.write(json_string)
    # outfile.close()
    gameind+=1
```

```python
# import seaborn as sns
# sns.set()
labels, data = [*zip(*jsondict.items())]
plt.boxplot(data)
plt.xticks(range(1, len(labels) + 1), labels,rotation=90)
plt.ylabel("Motion Sickness Ratings")
plt.show()
```

```python
import csv
import matplotlib.pylab as plt
import pandas as pd
def GetObjectFrames(filename):
    numobjects_perframe = []
    df = pd.read_csv(filename)
    frames_col = df['framecounter']
    unique_framenumbers = list(set(frames_col))
    print(filename)
    for framenumber in unique_framenumbers:
        
        numobjects_perframe.append(list(frames_col).count(framenumber))
    return numobjects_perframe,unique_framenumbers

```

```python
def GetNumObjects(csvfile):
    selfreport_filename = csvfile.split('/')[-1].replace('.csv','')
    selfreport_filename_split = selfreport_filename.split(' ')
    gamename_sub = selfreport_filename_split[2]
    participant = selfreport_filename_split[1]
    # print(csvfile)
    # numobjects_perframe,frames_numobjects = GetObjectFrames(csvfile)
    for item in allfiles:
        if gamename_sub in item:
            if participant in item:
                if 'object.csv' in item:
                    numobjects_perframe,frames_numobjects = GetObjectFrames(item)
                    return numobjects_perframe,frames_numobjects
                
```

# Motion Sickness and Num Lights per frame

```python
import pandas as pd
def GetLightsPerFrame(csv):
    df = pd.read_csv(csv)
    frames = list(set(df['framecounter']))
    num_lights = []
    print(len(frames))
    lightnames_col = df['name']
    framecounter_col = df['framecounter']
    lightname_perframe = []
    for i in range(len(lightnames_col)):
        if i == 0 or framecounter_col[i] == framecounter_col[i-1]:
            lightname_perframe.append(lightnames_col[i])
        else:
            num_lights.append(len(lightname_perframe))
            lightname_perframe = []
        
    num_lights.append(len(lightname_perframe))
    print(len(num_lights))
    return num_lights,frames
```

```python
def GetLightIntensity(csvfile):
    selfreport_filename = csvfile.split('/')[-1].replace('.csv','')
    selfreport_filename_split = selfreport_filename.split(' ')
    gamename_sub = selfreport_filename_split[2]
    participant = selfreport_filename_split[1]
    for item in allfiles:
        if gamename_sub in item:
            if participant in item:
                if 'light.csv' in item:
                    lightintensity_perframe,frames_numobjects = GetLightsPerFrame(item)
                    return lightintensity_perframe,frames_numobjects
                
```

```python
import glob as glob
import os
import matplotlib.pylab as plt

specific_participant = '8 ' #'2 ' #'21 ' #
specific_game = 'Rome' #'Monster' #'Roller' #

mother = '../SelfReportData'
games = os.listdir('../DataVersion2')
if '.DS_Store' in games:
    games.remove('.DS_Store')
games = ['Beat Saber','Cartoon Network','Job Simulator', 'Monster Awakens', 'Traffic Cop', 'Voxel Shot','VR Rome','Epic Roller Coasters','Mini Motor Racing', 'Pottery']   
shapes = ['o','*','^','x','s','+']
fig, ax = plt.subplots()
gameind = 1
for game in games:   
    print(game)
    if specific_game not in game: continue
    game_scores = []
    # plt.subplot(3,3,gameind)
    
    i=0
    participants = []
    for report in glob.glob(mother+'/*.csv'):
        if game not in report: continue
        if specific_participant not in report: continue
        # print(report)
        scores,timestamps,frames_motionsickness = GetMotionSicknessScore(report,game)
        numobjects_perframe,frames_numobjects = GetLightIntensity(report)
        ax.plot(frames_motionsickness,scores,'r*-')
        ax2=ax.twinx()
        ax2.plot(frames_numobjects,numobjects_perframe,'b.')
        
        game_scores = game_scores+scores
        participants.append(report.split('/')[-1].split(' ')[0])
        i+=1
        plt.title('Game: '+game+', Participant: '+specific_participant)
    # plt.ylabel('Motion Sickness Level')
    # plt.xlabel('Reporting Instances')
    # ax.ylim(0,5)
    ax.set_ylabel("Motion Sickness Rating",color="red",fontsize=14)
    ax.set_xlabel('Frame Number')
    # ax2.ylim(0,max(numobjects_perframe))
    ax2.set_ylabel("Number of Light Elements",color="blue",fontsize=14)
    
    # plt.legend(participants)
    # print(game_scores)
    print(round(np.median(game_scores),2),round(np.mean(game_scores),2),round(np.std(game_scores),2),max(game_scores))
    gameind+=1
```
